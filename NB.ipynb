```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tennis = pd.read_csv('play_tennis.csv')
tennis
```

```python
inputs = tennis.drop('play',axis='columns')
target = tennis.play
```

```python
target
```

```python
dummies = pd.get_dummies(target)
dummies
```

```python
dummies1 = pd.get_dummies(inputs.outlook)
dummies2 = pd.get_dummies(inputs.temperature)
dummies3 = pd.get_dummies(inputs.humidity)
dummies4 = pd.get_dummies(inputs.wind)
dummies1
```

```python
dummies2
```

```python
dummies3
```

```python
dummies4
```

```python
inputs = pd.concat([inputs,dummies1,dummies2,dummies3,dummies4,dummies],axis='columns')
inputs
```

```python
inputs.drop(['outlook','temperature','humidity','wind'],axis='columns',inplace=True)
inputs
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
